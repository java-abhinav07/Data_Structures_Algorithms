Return subsets sum to K

Given an array A of size n and an integer K, return all subsets of A which sum to K.
Subsets are of length varying from 0 to n, that contain elements of the array. But the order of elements should remain same as in the input array.<br>
Note : The order of subsets are not important.

```python
def subsetsumkh(li,k,i):
    res1=[]
    res2=[]
    p=[]
    if li[i]==k:
        p.append([li[i]])
    if i<len(li)-1:
        res1=subsetsumkh(li,k,i+1)    
    if li[i]<k and i<len(li)-1:
        res2=subsetsumkh(li,k-li[i],i+1)
    for ele in res1:
        p.append(ele)    
    for ele in res2:
        p.append([li[i]]+ele)    
    return p

def subsetsumk(li,k,i):
    output=subsetsumkh(li,k,i)
    for i in range(len(output)):
        print(*output[i])
        
n=int(input())
li=[int(x) for x in input().split()]
k=int(input())
subsetsumk(li,k,0)
```

Return subset of an array
<br>
Given an integer array (of length n), find and return all the subsets of input array.<br>
Subsets are of length varying from 0 to n, that contain elements of the array. But the order of elements should remain same as in the input array.<br>
Note : The order of subsets are not important.<br>

Input format :

Line 1 : Size of array

Line 2 : Array elements (separated by space)


```python
def subset(arr, n):
    if n == 0:
        return [[]]
    elif n == 1:
        return [[arr[0]], []]

    subsets_excluding = subset(arr[1:], n-1)

    ans = subsets_excluding[:]
    for sub in subsets_excluding:
        sub2 = sub[:]
        sub2.insert(0, arr[0])
        ans.append(sub2)

    return ans



n = int(input())
arr = list(int(i) for i in input().split())
for sub in (subset(arr, n)):
    for element in sub:
        print(element, end = " ")
    print()
```

Return Permutations of a String
<br>
Given a string, find and return all the possible permutations of the input string.<br>
Note : The order of permutations are not important.


```python
# using library function
import itertools
def permutations(string):
    ans = []
    for element in itertools.permutations(string):
        x = ""
        for tup in element:
            x += str(tup)
        ans.append(x)
    return ans


string = input()
ans = permutations(string)
for s in ans:
    print(s)
```

```python
def permutations(string):
    if not string:
        return []
    elif len(string) == 1:
        return [string]
    ans = []
    
    for element in string:
        firstcharacter = element
        copy = string.replace(firstcharacter, '')
        y = permutations(copy)
        for item in y:
            ans.append(firstcharacter + item)
            
    return ans
        
        

string = input()
ans = permutations(string)
for s in ans:
    print(s)
```

Print Subsets of Array

Given an integer array (of length n), find and print all the subsets of input array.<br>
Subsets are of length varying from 0 to n, that contain elements of the array. But the order of elements should remain same as in the input array.<br>
Note : The order of subsets are not important. Just print the subsets in different lines.<br>

```python
def printsubsets(n, arr, ans = [[]]):
    if n == 0:
        for subset in ans:
            for element in subset:
                print(element, end=" ")
            print()
        return 
    
    newans = ans[:]
    for element in ans:
        ele = element[:]
        ele.append(arr[0])
        newans.append(ele)
        
    printsubsets(n-1, arr[1:], newans)
        

n = int(input())
arr = list(int(i) for i in input().split())
printsubsets(n, arr)
```
